# Стохастические градиентные методы


## Постановка задачи


Представим, что перед нами поставлена какая-нибудь задача машинного обучения: например, задача классификации. Для решения такой задачи мы предлагаем какую-нибудь модель $m$. Обозначим параметры модели $m$ через $x$. Также для оценки качества нашей модели мы конструируем функцию ошибки (loss function) $l$.


Допустим, существует некоторое распределение данных $\mathcal{D}$, из которого можно "получать" пары $(a, b)$, где $a$ - признаки, на которых наша модель должна научиться предсказывать некие правильные ответы $b$. Для успешного обучения модели, мы хотим минимизировать функцию потерь $l$ для **всех** данных из распределения $\mathcal{D}$. Другими словами

$$
\min_x [ f(x) = \mathbb{E}_{a,b \sim \mathcal{D}} [l(m(x,a),b)] ]
$$

Однако в реальности у нас в распоряжении ограниченный набор из $N$ пар вида $(a_i, b_i)$, а значит можно аппроксимировать $f(x)$ следующим образом

$$
f(x) \approx \frac1N \sum_{i=1}^N l(m(x,a_i),b_i)
$$

Обозначив $l(m(x,a_i),b_i)$ через $f_i(x)$, запишем исходную задачу машинного обучения в виде

$$
\min_x \frac1N \sum_{i=1}^N f_i(x).
$$

Обычно считается, что мы располагаем набором данных конечного размера, а значит справедливо следующее:

- Целевая функция является суммой **конечного** числа функций
- $N$ может быть очень большим


Применительно к прикладным проблемам, можно выделить следующие особенности получившейся задачи:

- Точное вычисление функции потерь,а значит и её градиента (следующий пункт), занимает очень много времени
- Минимизация отдельных $f_i(x)$ **не гарантирует** минимизации $\mathcal{f}(x)$!
- Высокая точность решения обычно не требуется
- Допустимо введение случайности


### Основная идея


$$
\begin{align*}
& f(x) = \frac1N \sum_{i=1}^N f_i(x) \\
& \nabla f(x) = \frac1N \sum_{i=1}^N \nabla f_i(x) \\
\end{align*}
$$


- Так как целевая функция равна сумме функций, то градиент равен сумме градиентов
- Можно вместо сложения всех градиентов, сложить только некоторые, выбранные некотором образом
- **Вопрос** - как выбрать, какие использовать?


## Разница между GD, SGD и MB SGD


### Градиентный спуск (GD)


$$
x_{k+1} = x_k - \gamma_k \frac1N \sum_{i=1}^{N} \nabla  f_i(x_k)
$$


На каждой итерации считаем полный градиент и делаем шаг в противоположном направлении. Иногда этот метод называют **полноградиентным**, противопоставляя его SGD и MB-SGD


**Гиперпараметры**

- $\gamma$, шаг обучения (learning rate)

**Плюсы**

- Точная оценка полного градиента
- Более стабильная сходимость

**Минусы**

- Дорого с точки зрения вычислений, особенно на больших датасетах
- За счёт вычисления полных градиентов, метод подвержен переобучению (overfitting)


#### **Сходимость** для **выпуклых** функций

Для шага $\gamma$ ([Theorem 3.4, Corollary 3.5](https://arxiv.org/pdf/2301.11235.pdf)):
$$f(x_k) - \inf f \leq \frac{\| x_0 - x^*\|^2}{2 \gamma k} = \mathcal{O}\left(\frac{1}{k}\right).$$

#### **Сходимость** для **сильно выпуклых** ($\mu$-convex) функций

Для шага $\gamma$ ([Theorem 3.6](https://arxiv.org/pdf/2301.11235.pdf)):
$$\|x_k-x^*\|^2 \leq (1-\gamma \mu)^k\| x_0 - x^*\|^2.$$

Другими словами, наблюдается линейная сходимость.

### Стохастический градиентный спуск (SGD)


$$ x_{k+1} = x_k - \gamma_k \nabla f_{i_k}(x_k) $$


На каждой итерации считаем градиент на основе одной случайно\* выбранной части данных.

**Вопрос**: случайно - это как?


**Гиперпараметры**

- $\gamma$, шаг обучения (learning rate)

**Плюсы**

- Вычислительно эффективный
- Возможность избежать локального минимума за счёт использования случайности

**Минусы**

- Заметные колебания в процессе сходимости
- Возможна более медленная сходимость на практике из-за грубой оценки полного градиента


#### **Сходимость** для **выпуклых** функций

Для начала обозначим **шум градиента** (gradient noise) как ([Definition 4.16](https://arxiv.org/pdf/2301.11235.pdf))
$$ \sigma \stackrel{def}{=} \inf_{x^* \in \argmin f} Var[\nabla f_i(x^*)]$$


Для фиксированного шага $\gamma$ ([Theorem 5.5](https://arxiv.org/pdf/2301.11235.pdf)):
$$\mathbb{E}[f(\bar{x}_k)- \inf f] \leq \frac{\| x_0 - x^*\|^2}{\gamma k} + 2\gamma \sigma.$$

Видно, что при фиксированном шаге алгоритм сходится к некоторой окрестности $2\gamma \sigma$.


Для убывающего шага $\gamma_k = \frac{\gamma_0}{\sqrt{k+1}}$ ([Theorem 5.7](https://arxiv.org/pdf/2301.11235.pdf)):
$$\mathbb{E}[f(\bar{x}_k)- \inf f] \leq \frac{5\| x_0 - x^*\|^2}{4\gamma_0 \sqrt{k}} + \gamma \sigma \frac{5\gamma_0 \log(k+1)}{\sqrt{k}} = \mathcal{O}\left(\frac{\log(k+1)}{{\sqrt{k}}}\right).$$



#### **Сходимость** для **сильно выпуклых** ($\mu$-convex) функций

Для фиксированного шага $\gamma$ ([Theorem 5.8](https://arxiv.org/pdf/2301.11235.pdf)):

$$\mathbb{E}\|x_k-x^*\|^2 \leq (1-\gamma \mu)^k\| x_0 - x^*\|^2 + \frac{2\gamma}{\mu} \sigma$$

Видно, что при фиксированном шаге алгоритм сходится к некоторой окрестности $\frac{2\gamma}{\mu}\sigma$.


[Theorem 5.1](https://arxiv.org/pdf/1907.04232.pdf) утверждает, что для убывающего шага оценка приобретает вид
$$\mathbb{E}\|x_k-x^*\|^2= \mathcal{O}\left(\frac{\ln k}{k}\right)$$

Однако делается оговорка, что для некоторых правил выбора шага, например для  $\gamma_k = \frac{1}{k}$, возможно получить оценку $\mathcal{O}\left(\frac{1}{k}\right)$.


В [статье](https://arxiv.org/pdf/1907.04232.pdf) показано, что оптимальная оценка достигается для убывающего шага вида $\gamma_k = \min\{\frac{1}{2L}, \frac{\ln(\max{\{ 2, \mu^2 \|x_k-x^*\|^2 k / \sigma^2\}})}{\mu k}\}$, и представляет собой
$$\mathbb{E}\|x_k-x^*\|^2 = \tilde{\mathcal{O}}\left(\|x_k-x^*\|^2 \exp \left[-\frac{\mu k}{2L}\right] + \frac{\sigma^2}{\mu^2k}\right).$$

### Mini Batch Stochastic Gradient Descent (MB SGD)


$$
x_{k+1} = x_k - \gamma_k \frac{1}{|\mathcal{X_k}|} \sum_{i \in \mathcal{X_k}} \nabla  f_i(x_k)
$$


Компромисс между GD и SGD - на каждой итерации градиент считается на основе $b=|\mathcal{X_k}|$ случайно\* выбранных частях данных.
$b$ называется размером батча.

На практике обычно используется именно такой метод. Поэтому когда ссылаются на SGD, имеется ввиду обычно MB SGD


**Гиперпараметры**

- $\gamma$, шаг обучения (learning rate)
- $b$, размер батча (batch size)

**Плюсы**

- Потенциальный компромисс между вычислительной эффективностью и стабильностью процесса сходимости

**Минусы**

- Необходимость подбора дополнительного гиперпараметра $b$

#### **Сходимость** для **выпуклых** функций

Будем считать **шум градиента для батча** (minibatch gradient noise) как ([Definition 6.4, Lemma 6.5](https://arxiv.org/pdf/2301.11235.pdf))
$$ \sigma_b = \frac{N - b}{b(N-1)} \sigma$$


Для фиксированного шага $\gamma$ ([Theorem 6.9](https://arxiv.org/pdf/2301.11235.pdf)):
$$\mathbb{E}[f(\bar{x}_k)- \inf f] \leq \frac{\| x_0 - x^*\|^2}{\gamma k} + 2\gamma \sigma_b.$$

Видно, что при фиксированном шаге алгоритм сходится к некоторой окрестности $2\gamma \sigma_b$.


Для убывающего шага $\gamma_k = \frac{\gamma_0}{\sqrt{k+1}}$ ([Theorem 6.11](https://arxiv.org/pdf/2301.11235.pdf)):
$$\mathbb{E}[f(\bar{x}_k)- \inf f] \leq \frac{5\| x_0 - x^*\|^2}{4\gamma_0 \sqrt{k}} + \gamma \sigma_b \frac{5\gamma_0 \log(k+1)}{\sqrt{k}} = \mathcal{O}\left(\frac{\log(k+1)}{{\sqrt{k}}}\right).$$


Заметим, что теоремы напоминают соответствующие теоремы для SGD. Действительно, единственное изменение - смена $\sigma$ на $\sigma_b$. А значит, в случае MB SGD появляется зависимость скорости сходимости от размера батча.


#### **Сходимость** для **сильно выпуклых** ($\mu$-convex) функций

Для фиксированного шага $\gamma$ ([Theorem 6.12](https://arxiv.org/pdf/2301.11235.pdf)):
$$\mathbb{E}\|x_k-x^*\|^2 \leq (1-\gamma \mu)^k\| x_0 - x^*\|^2 + \frac{2\gamma}{\mu} \sigma_b.$$

Видно, что при фиксированном шаге алгоритм сходится к некоторой окрестности $\frac{2\gamma}{\mu}\sigma_b$.

Аналогично SGD, используя убывающий шаг можно добиться лучших оценок: $\mathcal{O}\left(\frac{\ln k}{k}\right)$, и даже $\mathcal{O}\left(\frac{1}{k}\right)$.  Например, для убывающего шага $\gamma_k = \min\{\frac{1}{2L}, \frac{\ln(\max{\{ 2, \mu^2 \|x_k-x^*\|^2 k / \sigma_b^2\}})}{\mu k}\}$:
$$\mathbb{E}\|x_k-x^*\|^2 = \tilde{\mathcal{O}}\left(\|x_k-x^*\|^2 \exp \left[-\frac{\mu k}{2L}\right] + \frac{\sigma_b^2}{\mu^2k}\right).$$

## Техники практического улучшения SGD


### Техники батчирования


#### Независимое батчирование или батчирование с повторениями (Independent sampling)


Такой подход обычно используется в теории, так как проще поддаётся анализу. Во время формирования батчей, каждый образце данных может быть включён в несколько батчей (с повторениями).

При такой технике, minibatch gradient noise имеет значение $\sigma_b = \frac{1}{b} \sigma$.

**Плюсы**

- Возможность переиспользования одних и тех же данных в разных батчах
- Более разнообразнын обновления

**Минусы**

- Может привести к шуму из-за повторяющихся данных
- Требует аккуратного тюннинга стратегии для выбора с повторениями


#### Батчирование без повторений (Partition sampling)


Такой подход наиболее распространён на практике. Каждый образец данных используется ровно один раз в рамках одной эпохи. Другими словами, образцы данных в батчах не пересекаются, а значит за одну эпоху (один проход по всем батчам) алгоритм использует все доступные данные.


При такой технике, minibatch gradient noise имеет значение $\sigma_b = \frac{N - b}{b(N-1)} \sigma$.


**Плюсы**

- Нет избыточности, так как каждый образец данных используется в рамках эпохи только один раз

**Минусы**

- Может привести к менее разнообразным обновлениям, особенно для небольших датасетов


#### Другие техники

Наряду с  Independent sampling и Partition sampling в [литературе](https://proceedings.mlr.press/v97/qian19b/qian19b.pdf) можно встретить следующие техники батчирования:

- **Single element sampling.** Только множества, состоящие из одного экземпляра данных, имеют ненулевой шанс быть выбранными в качестве батча
- **$\tau$-nice sampling.** Батч формируется равновероятностным выбором из всех подмножеств датасета $\mathcal{D}$ размером $\tau$

Видно, что Single element sampling эквивалентен случаю с Independent Sampling и размером батча равному 1, а $\tau$-nice sampling - случаю с Independent Sampling и размером батча равному $\tau$. Поэтому отдельно эти техники реализованы в данном ноутбуке не будут.

### Отсечение градиента (Gradient clipping)


Gradient clipping используется для смягчения exploding gradient problem. Exploding gradient problem возникает, когда градиенты во время обучения становятся слишком большими, что приводит к нестабильности и трудностям в сходимости. Clipping устраняет эту проблему путем уменьшения градиентов, если их норма превышает заданный порог. Обычно используется $L_1$ или $L_2$ норма.

В случае SGD это означает, что мы не предотвращаем ситуацию, когда градиент одних батчей влияет намного сильнее, чем градиент других.

Например, для порогового значения $T$ и градиента $\nabla \theta$, формула для gradient clipping выглядит следующим образом

$$
\nabla_{clipped} = \frac{\text{T}}{\max(\text{T},\| \nabla \theta \| )} \nabla \theta
$$

В случае $L_2$ нормы, gradient clipping эквивалентен проекции градиента на шар.


**Плюсы**

- Уменьшает скачкообразные градиенты: предотвращает чрезмерное увеличение градиентов во время тренировки, снижая рист числовой нестабильности
- Позволяет поддерживать более стабильный тренировочный процесс

**Минусы**

- Потеря информации, так как изменяет непосредственно значение градиента
- Тюнинг дополнительно гиперпараметра


### Техники перемешивание данных


#### Random Reshuffling (RR)


На каждой эпохе, RR генерирует новую случайную перестановку данных, используя которую делаются итерации градиентного спуска.

$$
\begin{align*}
\textbf{Input:} & \text{ Stepsize } \gamma > 0, \text{ initial vector } x_0 = x_0^0 \in \mathbb{R}^d, \text{ number of epochs } T. \\
1: & \text{for epochs } t = 0, 1, \ldots, T - 1 \text{ do} \\
2: & \quad \text{Sample a permutation } \{\pi_0, \pi_1, \ldots, \pi_{n-1}\} \text{ of } \{1, 2, \ldots, n\} \\
3: & \quad \text{for } i = 0, 1, \ldots, n - 1 \text{ do} \\
4: & \quad \quad x_{i+1}^t = x_i^t - \gamma \nabla f_{\pi_i}(x_i^t) \\
5: & \quad x_{t+1} = x_n^t \\
\end{align*}
$$


**Плюсы**

- Хорошая генерализация, так как модель не учится запоминать порядок тренировочных данных
- Гарантирует, что батчи хорошо представляют распределение данных
- Элемент случайности может помочь в преодолении локального минимума

**Минусы**

- Требует дополнительные вычислительные мощности на генерацию новых перестановок на каждой эпохе


#### Shuffle-Once (SO)


Так же как и RR, SO перемешивает датасет, но делает это только в начале тренировки. Полученная перестановка используется во всех последующих эпохах.

$$
\begin{align*}
\textbf{Input:} & \text{ Stepsize } \gamma > 0, \text{ initial vector } x_0 = x_0^0 \in \mathbb{R}^d, \text{ number of epochs } T. \\
1: & \text{Sample a permutation } \{\pi_0, \pi_1, \ldots, \pi_{n-1}\} \text{ of } \{1, 2, \ldots, n\} \\
2: & \text{for epochs } t = 0, 1, \ldots, T - 1 \text{ do} \\
3: & \quad \text{for } i = 0, 1, \ldots, n - 1 \text{ do} \\
4: & \quad \quad x_{i+1}^t = x_i^t - \gamma \nabla f_{\pi_i}(x_i^t) \\
5: & \quad x_{t+1} = x_n^t \\
\end{align*}
$$



**Плюсы**

- Не требует дополнительных затрат мощности, так как перестановка генерируется только один раз

**Минусы**

- Потенциально модель может переобучиться на определённом порядке тренировочных данных
- Может сложится ситуация, когда некоторые батчи не являются репрезентативными на всём периоде обучения


#### Incremental Gradient (IG)

Техника идентична SO, за исключением того, что начальная перестановка является не случайной, а детерминированной. Следовательно, IG выполняет постепенные градиентные шаги по данным *циклическим* образом. Порядок может быть произвольным, например, он может быть выбран *неявно* в соответствии с порядком поступления данных или выбран *случайным* образом. Наиболее часто используется на практике.

Концептуальное отличие IG от SO заключается в том, что без перемешивания (IG) данные могут лежать неоднородно, в отличие от случая, когда перемешивание происходит хотя бы раз (SO).

**Плюсы**

- Не требует дополнительных затрат мощности, так как перестановка генерируется только один раз

**Минусы**

- Потенциально модель может переобучиться на определённом порядке тренировочных данных
- Может сложится ситуация, когда некоторые батчи не являются репрезентативными на всём периоде обучения

### Вспомогательные функции


```python
def train(
    data,
    model: nn.Module,
    optimizer: torch.optim.Optimizer,
    loss_fn,
    device,
):
    inputs, labels = data
    inputs, labels = (
        inputs.to(device),
        labels.to(device),
    )

    # zero the parameter gradients
    optimizer.zero_grad()

    # forward pass and loss calculation
    outputs = model(inputs)
    return loss_fn(outputs, labels)


def eval(
    test_loader: DataLoader,
    model: nn.Module,
    loss_fn,
    device,
):
    correct = 0
    total = 0
    val_loss = 0.0
    with torch.no_grad():
        model.eval()  # evaluation mode
        for data in test_loader:
            inputs, labels = data
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            val_loss += loss_fn(outputs, labels).item()

    return val_loss, correct / total


def train_eval_loop(
    model: nn.Module,
    optimizer: torch.optim.Optimizer,
    loss_fn,
    train_loader: DataLoader,
    test_loader: DataLoader,
    train_fn,
    eval_fn,
    configs: dict,
    device,
    log_frequency: int,
    verbose: bool,
    include_starting_point: bool,
    leave_tqdm: bool,
    epochs=1,
):
    loss_logs = []
    val_loss_logs = []
    time_logs = []
    iterations_logs = []
    val_metric_logs = []

    torch.cuda.synchronize()
    start_time = time.time()

    iterations = 0

    if include_starting_point:
        # logs
        iterations_logs.append(iterations)
        time_logs.append(0)

        # evaluation
        val_loss, val_metric = eval_fn(test_loader, model, loss_fn, device=device)

        val_loss_logs.append(val_loss)
        val_metric_logs.append(val_metric)

    # iterating over epochs
    for epoch in range(epochs):
        # training loop description
        train_loop = tqdm(
            enumerate(train_loader),
            total=len(train_loader),
            desc=f"Epoch {epoch+1}",
            disable=not verbose,
            leave=leave_tqdm,
        )
        train_loss = 0.0
        # iterate over dataset
        for i, data in train_loop:
            model.train()

            loss = train_fn(data, model, optimizer, loss_fn, device=device)

            # backward pass
            loss.backward()

            # gradient clipping
            if configs.get("clipping") and configs["clipping"]["apply"]:
                nn.utils.clip_grad_norm_(
                    model.parameters(),
                    max_norm=configs["clipping"]["threshold"],
                    norm_type=configs["clipping"]["norm_type"],
                )

            # optimizer run
            optimizer.step()

            train_loss += loss.item()

            iterations += 1
            if iterations % log_frequency != 0:
                continue

            # logs
            iterations_logs.append(iterations)
            loss_logs.append(loss.item())
            torch.cuda.synchronize()
            time_logs.append(time.time() - start_time)

            # evaluation
            val_loss, val_metric = eval_fn(test_loader, model, loss_fn, device=device)

            val_loss_logs.append(val_loss)
            val_metric_logs.append(val_metric)

            train_loop.set_postfix({"val loss": val_loss, "val acc": val_metric})

    return val_metric_logs, val_loss_logs, time_logs, iterations_logs
```

Функция для построения графиков


```python
PREDEFINED_COLORS = [
    "#ffa500",
    "#c83cbc",
    "#1c1c84",
    "#ff0000",
    "#08a4a7",
    "#008000",
]


def draw_method_plots(
    methods: list[dict],
    title: str,
    plots: list[
        tuple[dict, dict]
    ],  # 2 dicts for axes description, tuple for methods and values
    ylim=None,
    row_plots: int = 1,
    plot_width: float = 12,  # 16
    plot_height: float = 7,  # 7
    use_rainbow: bool = False,
    use_common_legend: bool = True,
    adjust: bool = False,
):
    num_plots = len(plots)
    if use_rainbow:
        num_colors = len(methods)
        cm = plt.get_cmap("gist_rainbow")
        colors = [cm(1.0 * i / num_colors) for i in range(num_colors)]
    else:
        colors = PREDEFINED_COLORS

    style_cycler = cycler(linestyle=["-", "--", ":", "-."]) * cycler(color=colors)
    column_plots = math.ceil(num_plots / row_plots)

    fig, axs = plt.subplots(
        column_plots,
        row_plots,
        figsize=(plot_width * row_plots, plot_height * column_plots),
    )
    fig.suptitle(title, fontsize=14)
    axs_list = [axs] if num_plots == 1 else list(axs.flat)

    for ax in axs_list:
        ax.grid()
        ax.set_prop_cycle(style_cycler)
        if ylim is not None:
            ax.set_ylim(top=ylim)

    for ax, (p1, p2) in zip(axs_list, plots):
        ax.set_title(f"{p2['name']} over {p1['name']}")
        ax.set(xlabel=p1["label"], ylabel=p2["label"])

        if p1.get("log", False):
            ax.set_xscale("log")
        if p2.get("log", False):
            ax.set_yscale("log")

        for method in methods:
            label = method["name"]

            x_values = method[p1["name"]]
            y_values = method[p2["name"]]

            ax.plot(x_values, y_values, label=label)
            ax.scatter(x_values[-1], y_values[-1], s=15)

    if use_common_legend:
        lines_labels = [axs_list[0].get_legend_handles_labels()]
        lines, labels = [sum(x, []) for x in zip(*lines_labels)]
        fig.legend(
            lines,
            labels,
            scatterpoints=1,
            markerscale=3,
            loc="outside lower center",
            ncol=min(6, len(methods)),
            bbox_to_anchor=(0.5, -0.05),
        )
    else:
        if len(methods) > 1:
            for ax in axs_list:
                ax.legend()

    plt.tight_layout()
    if adjust:
        plt.subplots_adjust(
            top=1 - 0.1 / (num_plots**0.5), bottom=0.12 / (num_plots**2), hspace=0.15
        )
    plt.show()
```

```python
from copy import deepcopy

from torch.utils.data import Subset


def shuffle_dataset(dataset):
  return Subset(dataset, np.random.permutation(np.arange(len(dataset))))

def choose_train_loader(sequence, batch_size, configs: dict, **kwargs) -> DataLoader:
    shuffle = False
    train_seq = deepcopy(sequence)

    shuffling = configs.get("shuffling", None)
    if shuffling == "RR":
        shuffle = True
    elif shuffling == "SO":
        train_seq = shuffle_dataset(train_seq)

    sampler = (
        RandomSampler(train_seq, replacement=True)
        if configs.get("replacement", False)
        else None
    )

    return DataLoader(train_seq, batch_size, shuffle=shuffle, sampler=sampler, **kwargs)


def methods_run(
    model_class,
    loss_fn,
    train_loader_lambda,
    test_loader: DataLoader,
    to_compare: dict[str, dict],
    epochs: int = 5,
    log_frequency: float = 50,
    seed: int = 420,
    verbose: bool = False,
    train_fn=train,
    eval_fn=eval,
    include_starting_point: bool = False,
    device=DEVICE,
    leave_tqdm: bool = False,
) -> list[dict]:

    comparison_dicts = []

    for name, configs in to_compare.items():
        if verbose:
            print(f"Running {name.upper()}...")

        torch.manual_seed(seed)
        torch.cuda.manual_seed(seed)
        np.random.seed(seed)

        model = model_class().to(device)
        optimizer = configs["optimizer_lambda"](model)
        train_loader = train_loader_lambda(configs)

        if log_frequency <= 0:
            literal_log_frequency = len(train_loader)
        elif log_frequency < 1:
            literal_log_frequency = int(len(train_loader) * log_frequency)
        else:
            literal_log_frequency = int(log_frequency)

        val_acc_logs, val_loss_logs, time_logs, iterations_logs = train_eval_loop(
            model,
            optimizer,
            loss_fn,
            train_fn=train_fn,
            eval_fn=eval_fn,
            epochs=epochs,
            train_loader=train_loader,
            test_loader=test_loader,
            configs=configs,
            device=device,
            log_frequency=literal_log_frequency,
            verbose=verbose,
            include_starting_point=include_starting_point,
            leave_tqdm=leave_tqdm,
        )

        comparison_dicts.append(
            {
                "name": name,
                "Test accuracy": val_acc_logs,
                "Test loss": val_loss_logs,
                "Time": time_logs,
                "Iterations": iterations_logs,
            }
        )

    if verbose:
        print("Done!")
    return comparison_dicts


def methods_plot_comparison(title: str, comparison_dicts: list[dict], only_loss: bool = False):

    plots = [            (
                {"name": "Iterations", "label": "Iterations"},
                {"name": "Test loss", "label": "Loss"},
            ),
            (
                {"name": "Time", "label": "Running time, seconds"},
                {"name": "Test loss", "label": "Loss"},
            ),
        ]
    if not only_loss:
        plots =[(
                {"name": "Iterations", "label": "Iterations"},
                {"name": "Test accuracy", "label": "Accuracy"},
            ),
            (
                {"name": "Time", "label": "Running time, seconds"},
                {"name": "Test accuracy", "label": "Accuracy"},
            )] + plots

    draw_method_plots(
        comparison_dicts,
        title,
        plots,
        plot_width=7,
        plot_height=4,
        row_plots=2,
    )
```

## Модификации SGD


### SGD

$$
\begin{align*}
& x_{k+1} = x_k - \gamma_k \sum_{i\in \mathcal{X}} \nabla f_i(x_k) \\
\end{align*}
$$


Классический метод оптимизации, широко используемый в машинном обучении. SGD является одним из самых популярных и простых методов численной оптимизации. Теоретически, сходимость SGD является оптимальной, однако на практике различные модификации показывают себя намного лучше, что будет видно в предстоящих экспериментах.


**Гиперпараметры**

- $\gamma$, шаг обучения (learning rate)

Здесь и далее, не включаем в список гиперпараметров размер батча $b$


**Плюсы**

- Простота тюннинга, благодаря небольшому количеству гиперпараметров
- Простота реализации

**Минусы**

- Относительная (!!!) чувствительность к выбору $\gamma$
- Отсутствие адапитвного выбора шага (шкалирования градиентов)
- Может быть менее надежным при наличии зашумленных градиентов, поскольку он полагается на точную информацию о градиенте
- Возможны заметные колебания в процессе сходимости


**Когда использовать?**

- В относительно простых (низкоразмерных) задачах, когда модель не очень сложна и дополнительная сложность оптимизаторов, таких как Adam или RMSProp, может быть просто избыточными
- В ситуациях, когда дополнительные вычисления дорогие
- Когда есть необходимость точно и вручную регулировать гиперпараметры


### Метод тяжёлого шарика (Heavy Ball)

В теории обычно записывается как

$$
\begin{align*}
& x_{k+1} = x_k - \gamma_k \sum_{i\in \mathcal{X}} \nabla f_i(x_k) + \beta_k (x_k - x_{k-1})
\end{align*}
$$

Однако на практике обычно используется следующая запись:

$$
\begin{align*}
& h_k = \beta_k h_{k-1} - \gamma_k \sum_{i\in \mathcal{X}} \nabla f_i(x_k) \\
& x_{k+1} = x_k + h_k
\end{align*}
$$

Метод тяжёлого шарика был предложен в 1964 году Борисом Поляком. Это классическое улучшение стандартного SGD за счёт добавления импульса (моментума): метод комбинирует текущий градиент с предыдущей историей для ускорения сходимости. Мы как бы предполагаем наличие у текущей точки есть инерция, как у "тяжелого" материального шарика движущегося по наклонной поверхности


**Гиперпараметры**

- $\gamma$, шаг обучения (learning rate)
- $\beta$, импульс (momentum). Определяет силу влияния предыдущих шагов на текущее обновление. Выбирается из диапазона $[0,1]$. Обычно близок к 1, например 0.99

**Плюсы**

- Понятная физика и интуиция
- Простота реализации
- Дешевизна вычислений
- Ускорение движения вдали от оптимума
- Возможность избегания локального минимума за счёт "инерции"

**Минусы**

- Необходимость подбора двух гиперпараметров (вместо одного в случае SGD)
- Отсутствие адапитвного выбора шага (шкалирования градиентов)
- Относительная (!!!) чувствительность к значению $\beta$


**Когда использовать?**

- В относительно простых (низкоразмерных) задачах, когда модель не очень сложна и дополнительная сложность оптимизаторов, таких как Adam или RMSProp, может быть просто избыточными
- В задачах с несколькими локальными минимумами
- В ситуациях, когда дополнительные вычисления дорогие
- Когда требуется стабильная сходимость. В определённых сценариях, более сложные оптимизаторы могут приводить к колебаниям или нестабильности
- Когда есть необходимость точно и вручную регулировать гиперпараметры


### Ускоренный градиентный метод (Нестерова)



В теории обычно записывается как

$$
\begin{align*}
& x_{k+1} = y_k - \gamma_k \sum_{i\in \mathcal{X}} \nabla f_i(y_k) \\
& y_{k+1} = x_{k+1} + \beta_k (x_{k+1} - x_k)
\end{align*}
$$

Однако на практике обычно используется следующая запись:

$$
\begin{align*}
& h_k = \beta_k h_{k-1} - \gamma_k \sum_{i\in \mathcal{X}} \nabla f_i(x_k + \beta_k  h_{k-1})\\
& x_{k+1} = x_k + h_k
\end{align*}
$$


Метод Нестерова - это алгоритм оптимизации, который расширяет идею импульса в SGD. Был представлен Юрием Нестеровым в 1983 году. Основная идея заключается в использовании частичного прогнозирования при обновлении параметров, включая импульс для ускорения сходимости. Ключевое отличие от метода тяжёлого шарика в том, что благодаря импульсу в точку подсчёта градиента, происходит экстраполяция, или "взгляд в будущее"


**Гиперпараметры**

- $\gamma$, шаг (learning rate)
- $\beta$, импульс (momentum) и импульс Нестерова (Nesterov momentum). Обычно (как и в реализации PyTorch) оба импульса выбираются одинаковыми. Моментум определяет объем предыдущего обновления, который переносится на текущее обновление. Моментум Нестерова определяет, как сильно алгоритм "смотрит вперед" на направление, в котором движется алгоритм оптимизации, чтобы использовать эту информацию для настройки размера шага и направления. Обычно близок к 1, например 0.99


**Плюсы**

- Понятная физика и интуиция
- Простота реализации
- Дешевизна вычислений
- Ускорение движения вдали от оптимума
- Возможность избегания локального минимума за счёт импульса

**Минусы**

- Более вычислительно сложный, по сравнению с SGD и Heavy Ball
- Необходимость подбора двух гиперпараметров (вместо одного в случае SGD)
- Отсутствие адапитвного выбора шага (шкалирования градиентов)
- Относительная (!!!) чувствительность к значению $\beta$


**Когда использовать?**

- В относительно простых (низкоразмерных) задачах, когда модель не очень сложна и дополнительная сложность оптимизаторов, таких как Adam или RMSProp, может быть просто избыточными
- В задачах и большой кривизной ландшафта
- В задаче с Ill-conditioned landscape - когда собственные значения Гессиана очень сильно варьируются. Когда некоторые собственные значения намного больше других, ландшафт оптимизации становится сильно вытянутым или растянутым в определенных направлениях. Моментум помогает эффективно ориентироваться в таких ситуациях
- В ситуациях, когда дополнительные вычисления дорогие
- Когда есть необходимость точно и вручную регулировать гиперпараметры


### AdaGrad

$$
\begin{align*}
& g_k =  \sum_{i\in \mathcal{X}} \nabla f_i(x_k)\\
& G_k = G_{k-1} + g_k^2 \\
& x_{k+1} = x_k - \frac{\gamma_k}{\sqrt{G_k}+ \epsilon} g_k
\end{align*}
$$


AdaGrad (Adaptive Gradient) - это алгоритм оптимизации, предшественник RMSProp, предназначенный для адаптивной настройки скорости обучения для различных параметров модели. Он был разработан для решения проблем, связанных с установкой единой скорости обучения для всех параметров, особенно в ситуациях, где признаки имеют значительно различающуюся частоту или важность: хорошо бы уметь обновлять параметры с оглядкой на то, насколько типичный признак они фиксируют.

Важный момент: обычно нотация $g_k^2$ используется для
сокращённой записи $\text{diag} (g_k \odot g_k)$. Мы будем придерживаться этой традиции здесь и далее в описаниях алгоритмов оптимизации.

Грубо говоря, суть AdaGrad в том, чтобы использовать *что-нибудь*, что бы уменьшало обновления для параметров, которые и так часто обновляются. В частности, из приведённого выше алгоритма AdaGrad, можно получить целое семейство алгоритмов, модифицируя формулы обновления истории ($G_k$) и шкалирования.
Например, можно попробовать накапливать не $g_k^2$, а $g_k^4$, или делить не на $\sqrt{G_k}$, а на что-нибудь вроде $G_k$ или $e^{G_k}$

**Параметры**

- $G_k$, накопление информации о предыдущих обновлениях. Это может быть сумма квадратов прошлых градиентов или, например, их модулей
- $\epsilon$, маленькая константа для обеспечения численной стабильности. Обычно порядка $10^{-8}$

**Гиперпараметры**

- $\gamma$, шаг обучения (learning rate)


**Плюсы**

- Адаптивный выбор шага,в зависимости от истории обновления
- Отсутствие необходимости точно подбирать скорость обучения. Достаточно выставить её в меру большой, чтобы обеспечить хороший запас, но не такой огромной, чтобы алгоритм расходился. По сути мы автоматически получаем затухание скорости обучения  (learning rate decay)
- Возможность выбора алгоритмов накопления истории и шкалирования

**Минусы**

- Старые градиенты влияют так же сильно, как и недавние
- $G_k$ может увеличиваться сколько угодно, что через некоторое время приводит к слишком маленьким обновлениям и стазису
- Дополнительные затраты памяти для хранения $G_k$
- Отсутствие импульса, что может быть критично при попадании в локальный оптимум
- Необходимость в экспериментировании для выбора оптимальных техник накопления истории и шкалирования


**Когда использовать?**

- Когда данные разрежены, например в обработке естественного языка (NLP) или рекомендательных системах, где частота и важность признаков сильно варьируются
- При онлайн-обучении. AgaGrad способен на ходу адаптироваться к изменениям распределения признаков
- Когда трудно настроить длину шага и/или схему его постепенного убывания
- В задачах с большим количеством признаков (многоразмерные задачи)


### RMSProp


$$
\begin{align*}
& g_k =  \sum_{i\in \mathcal{X}} \nabla f_i(x_k)\\
& G_k = \beta G_{k-1} + (1-\beta) g_k^2  \\
& x_{k+1} = x_k - \frac{\gamma_k}{\sqrt{G_k} + \epsilon} g_k
\end{align*}
$$


RMSProp (Root Mean Square Propagation) был предложен Джеффри Хинтоном в 2012 году для решения проблем адаптации длины шага во время обучения. Ключевая идея RMSProp заключается в масштабировании градиента каждого веса в модели путем деления его на среднеквадратичное значение градиентов этого веса. Это помогает предотвратить слишком быстрое обучение весов с высокими градиентами, в то же время позволяя весам с низкими градиентами продолжать обучение быстрее.

RMSProp решает основную проблему AdaGrad - старые градиенты и недавние градиенты имеют разные веса, то есть имеет место некое "забывание" истории


**Параметры**

- $G_k$, взвешенное скользящее среднее квадратичных градиентов. Инициализируется $G_0 = 0$
- $\epsilon$, маленькая константа для обеспечения численной стабильности. Обычно порядка $10^{-8}$

**Гиперпараметры**

- $\gamma$, шаг обучения (learning rate)
- $\beta$, импульс (momentum). Обычно близок к 1, например 0.99


**Плюсы**

- Хорошо справляется с зашумленными функциями
- Сглаживает проблемы слишком больших/маленьких градиентов (exploding/vanishing gradients issues)
- Обычно хорошо показывает себя и в невыпуклом случае

**Минусы**

- Довольно чувствительный к выбору моментума
- Смещённая оценка скользящего среднего может привести к суб-оптимальному решению


**Когда использовать?**

- Когда у модели с большим количеством параметров во время тренировке наблюдаются колебания, RMSProp может помочь стабилизировать процесс обучения, подстраиваясь под градиент
- Когда трудно настроить длину шага. RMSProp может быть эффективным, поскольку он масштабирует градиенты, что может помочь процессу оптимизации проходить более плавно независимо от длины шага
- Когда целевая функция зашумлена, RMSProp может сгладить краткосрочные колебания и выделить долгосрочные тенденции, которые помогают смягчить воздействие шума и позволяют модели быстрее сходиться
- Когда требуется или есть возможность сохранить баланс между вычислительными затратами и адаптивностью


### Adam


$$
\begin{align*}
& g_k =  \sum_{i\in \mathcal{X}} \nabla f_i(x_k)\\
& m_k = \beta_1 m_{k-1} + (1-\beta_1) g_k  \\
& v_k = \beta_2 v_{k-1} + (1-\beta_2) g_k^2  \\
& \hat{m_k} = \frac{m_k}{1-\beta_1^k} \\
& \hat{v_k} = \frac{v_k}{1-\beta_2^k} \\
& x_{k+1} = x_k - \frac{\gamma_k}{\sqrt{\hat{v_k}} + \epsilon} \hat{m_k}
\end{align*}
$$


Adam (Adaptive Moment Estimation) был представлен в статье 2015 года [Adam: A Method for Stochastic Optimization](https://arxiv.org/abs/1412.6980). Adam поддерживает экспоненциальные скользящие средние весов и градиентов, которые он использует для масштабирования скорости обучения. Другими словами, Adam использует оценки среднего значения и дисперсии градиентов для адаптивного масштабирования скорости обучения во время тренировки, что может повысить скорость и стабильность процесса оптимизации.

Полезно бывает рассмотреть следующий вид формулы для коррекции ($\hat{v_k}$):

$$
\hat{v_k} = \frac{(1-\beta_2)\sum_{i=1}^k \beta_2^{k-i} \text{diag}(g_i \odot g_i)}{1 - \beta_2^k}.
$$

Идея в том, что $g_k^2$ берётся с весом $1$, $g_{k-1}^2$ берётся с весом $\beta_2$, $g_{k-2}^2$ берётся с весом $\beta_2^2$, и так далее.
Далее все веса делятся на $1+\beta_2+\beta_2^2+\dots+\beta_2^k = \frac{1 - \beta_2^k}{(1-\beta_2)}$. Таким образом, сумма всех весов равна 1, то есть получилась выпуклая комбинация.
Таким образом, веса для градиентов зависят от номера итерации. На начальных итерация ведёт себя похожим на AdaGrad образом, а на поздних становится похожим на RMSProp.


**Параметры**

- $\hat{m}$ и $\hat{v}$, скорректированные с учетом смещения оценки первого и второго моментов градиентов. Инициализируются 0
- $\epsilon$, маленькая константа для обеспечения численной стабильности. Обычно порядка $10^{-8}$

**Гиперпараметры**

- $\gamma$, шаг обучения (learning rate)
- $\beta_1$, скорость затухания для первого момента. Обычно близок к 1, например 0.99
- $\beta_2$, скорость затухания для второго момента. Обычно близок к 1, например 0.999


**Плюсы**

- Отлично работает на практике
- Сглаживает проблемы слишком больших/маленьких градиентов (exploding/vanishing gradients issues)
- Эффективное использование памяти
- Относительно (!!!) нечувствителен к выбору гиперпараметров

**Минусы**

- Подразумевает больший объём вычислений, чем SGD или RMSProp
- Оценки первого и второго моментов смещены к 0 на ранних этапах обучения, что может привести к суб-оптимальному решению


**Когда использовать?**

- Когда нет возможности провести детальный тюннинг гиперпараметров
- В случае разряженных или зашумлённых градиентов, благодаря адаптивному подбору шага
- На больших датасетах: в таких случаем адаптивный подбор шага может быть особенно продуктивным


### Собственный PyTorch Optimizer


Реализуем простой метод с моментумом - Heavy Ball

```python
class CustomHeavyBall(
    torch.optim.Optimizer
):  # обязательно наследуемся от torch.optim.Optimizer
    def __init__(self, params, lr=1e-3, momentum=0.9):
        super(CustomHeavyBall, self).__init__(params, defaults={"lr": lr})

        self.momentum = momentum  # сохраняем гиперпараметр в переменную класса

        for group in self.param_groups:
            for p in group["params"]:
                # для каждого набора весов (параметров) инициализируем предыдущее направление
                self.state[p] = {"prev_momentum": torch.zeros_like(p.data)}

    @torch.no_grad()
    def step(self):
        # итерируемся по всем весам
        for group in self.param_groups:
            for p in group["params"]:
                if p.grad is None:
                    continue
                if p not in self.state:
                    self.state[p] = {"prev_momentum": torch.zeros_like(p.data)}

                # достаём из словаря состояний предыдущее направление для нужного параметра
                prev_momentum = self.state[p]["prev_momentum"]

                # Обновляем предыдущие параметры
                self.state[p]["prev_momentum"] =  self.momentum * prev_momentum - group["lr"] * p.grad.data

                # обновляем веса согласно Heavy Ball алгоритму
                p.data += self.state[p]["prev_momentum"]
```

```python
from torchtext.data.functional import to_map_style_dataset
from torchtext.data.utils import get_tokenizer
from torchtext.datasets import AG_NEWS
from torchtext.vocab import build_vocab_from_iterator
```

Загружаем датасет


```python
tokenizer = get_tokenizer("basic_english")

ag_test_batch_size = 128
ag_train_dataset = to_map_style_dataset(AG_NEWS(root="./data", split="train"))
ag_test_dataset = to_map_style_dataset(AG_NEWS(root="./data", split="test"))

ag_num_class = len(set([label for (label, text) in ag_train_dataset]))

print(f"{ag_num_class=}")
print(next(iter(ag_test_dataset)))


def yield_tokens(data_iter):
    for _, text in data_iter:
        yield tokenizer(text)


vocab = build_vocab_from_iterator(yield_tokens(ag_train_dataset), specials=["<unk>"])
vocab.set_default_index(vocab["<unk>"])

text_pipeline = lambda x: vocab(tokenizer(x))
label_pipeline = lambda x: int(x) - 1


def collate_batch(batch):
    label_list, text_list, offsets = [], [], [0]
    for _label, _text in batch:
        label_list.append(label_pipeline(_label))
        processed_text = torch.tensor(text_pipeline(_text), dtype=torch.int64)
        text_list.append(processed_text)
        offsets.append(processed_text.size(0))
    label_list = torch.tensor(label_list, dtype=torch.int64)
    offsets = torch.tensor(offsets[:-1]).cumsum(dim=0)
    text_list = torch.cat(text_list)
    return label_list.to(DEVICE), text_list.to(DEVICE), offsets.to(DEVICE)


ag_test_loader = DataLoader(
    ag_test_dataset,
    batch_size=ag_test_batch_size,
    shuffle=False,
    collate_fn=collate_batch,
)
```

Модель для классификации текста


```python
class AgNet(nn.Module):
    def __init__(self, vocab_size=len(vocab), embed_dim=64, num_class=ag_num_class):
        super(AgNet, self).__init__()
        self.embedding = nn.EmbeddingBag(vocab_size, embed_dim, sparse=False)
        self.fc = nn.Linear(embed_dim, num_class)
        self.init_weights()

    def init_weights(self):
        initrange = 0.5
        self.embedding.weight.data.uniform_(-initrange, initrange)
        self.fc.weight.data.uniform_(-initrange, initrange)
        self.fc.bias.data.zero_()

    def forward(self, text, offsets):
        embedded = self.embedding(text, offsets)
        return self.fc(embedded)
```

Новые функции для цикла тренировки, потому что сейчас нужно обрабатывать ещё и offsets


```python
def train_nlp(
    data,
    model: nn.Module,
    optimizer: torch.optim.Optimizer,
    loss_fn,
    device,
):
    labels, inputs, offsets = data
    labels, inputs, offsets = (
        labels.to(device),
        inputs.to(device),
        offsets.to(device),
    )

    # zero the parameter gradients
    optimizer.zero_grad()

    # forward pass and loss calculation
    outputs = model(inputs, offsets)
    return loss_fn(outputs, labels)


def eval_nlp(
    test_loader: DataLoader,
    model: nn.Module,
    loss_fn,
    device,
):
    correct = 0
    total = 0
    val_loss = 0.0
    with torch.no_grad():
        model.eval()  # evaluation mode
        for labels, inputs, offsets in test_loader:
            inputs, labels, offsets = (
                inputs.to(device),
                labels.to(device),
                offsets.to(device),
            )

            outputs = model(inputs, offsets)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            val_loss += loss_fn(outputs, labels).item()

    return val_loss, correct / total
```

Сравнение улучшений. Используем Adam для лучшей демонстрации: сходимость SGD в данной проблеме оставляет желать лучшего, что будет видно в следующем эксперименте

Скачаем и подготовим датасет

```python
import math
import warnings
from typing import Iterable, List

import torch
import torch.nn as nn
from torch import Tensor
from torch.nn import Transformer
from torch.nn.utils.rnn import pad_sequence
from torch.utils.data import DataLoader
from torchtext.data.utils import get_tokenizer
from torchtext.datasets import Multi30k, multi30k
from torchtext.vocab import build_vocab_from_iterator

warnings.filterwarnings("ignore")


# https://github.com/pytorch/text/issues/1756#issuecomment-1163664163
multi30k.URL["train"] = "https://raw.githubusercontent.com/neychev/small_DL_repo/master/datasets/Multi30k/training.tar.gz"
multi30k.URL["valid"] = "https://raw.githubusercontent.com/neychev/small_DL_repo/master/datasets/Multi30k/validation.tar.gz"

SRC_LANGUAGE = "de"
TGT_LANGUAGE = "en"

token_transform = {}
vocab_transform = {}

token_transform[SRC_LANGUAGE] = get_tokenizer("spacy", language="de_core_news_sm")
token_transform[TGT_LANGUAGE] = get_tokenizer("spacy", language="en_core_web_sm")


def yield_tokens(data_iter: Iterable, language: str) -> List[str]:
    language_index = {SRC_LANGUAGE: 0, TGT_LANGUAGE: 1}

    for data_sample in data_iter:
        yield token_transform[language](data_sample[language_index[language]])

UNK_IDX, PAD_IDX, BOS_IDX, EOS_IDX = 0, 1, 2, 3
special_symbols = ["<unk>", "<pad>", "<bos>", "<eos>"]

for lg in [SRC_LANGUAGE, TGT_LANGUAGE]:
    train_iter = Multi30k(root="./data", split="train", language_pair=(SRC_LANGUAGE, TGT_LANGUAGE))
    vocab_transform[lg] = build_vocab_from_iterator(yield_tokens(train_iter, lg),
                                                    min_freq=1,
                                                    specials=special_symbols,
                                                    special_first=True)


for lg in [SRC_LANGUAGE, TGT_LANGUAGE]:
  vocab_transform[lg].set_default_index(UNK_IDX)

def generate_square_subsequent_mask(sz):
  mask = (torch.triu(torch.ones((sz, sz), device=DEVICE)) == 1).transpose(0, 1)
  mask = mask.float().masked_fill(mask == 0, float("-inf")).masked_fill(mask == 1, float(0.0))
  return mask


def create_mask(src, tgt):
    src_seq_len = src.shape[0]
    tgt_seq_len = tgt.shape[0]

    tgt_mask = generate_square_subsequent_mask(tgt_seq_len)
    src_mask = torch.zeros((src_seq_len, src_seq_len),device=DEVICE).type(torch.bool)

    src_padding_mask = (src == PAD_IDX).transpose(0, 1)
    tgt_padding_mask = (tgt == PAD_IDX).transpose(0, 1)
    return src_mask, tgt_mask, src_padding_mask, tgt_padding_mask


def sequential_transforms(*transforms):
    def func(txt_input):
        for transform in transforms:
            txt_input = transform(txt_input)
        return txt_input
    return func


def tensor_transform(token_ids: List[int]):
    return torch.cat((torch.tensor([BOS_IDX]),
                      torch.tensor(token_ids),
                      torch.tensor([EOS_IDX])))

text_transform = {}
for lg in [SRC_LANGUAGE, TGT_LANGUAGE]:
    text_transform[lg] = sequential_transforms(token_transform[lg],
                                               vocab_transform[lg],
                                               tensor_transform)

def multi_collate_fn(batch):
    src_batch, tgt_batch = [], []
    for src_sample, tgt_sample in batch:
        src_batch.append(text_transform[SRC_LANGUAGE](src_sample.rstrip("\n")))
        tgt_batch.append(text_transform[TGT_LANGUAGE](tgt_sample.rstrip("\n")))

    src_batch = pad_sequence(src_batch, padding_value=PAD_IDX)
    tgt_batch = pad_sequence(tgt_batch, padding_value=PAD_IDX)
    return src_batch, tgt_batch
```

Создание модели

```python
class PositionalEncoding(nn.Module):
    def __init__(self,
                 emb_size: int,
                 dropout: float,
                 maxlen: int = 5000):
        super(PositionalEncoding, self).__init__()
        den = torch.exp(- torch.arange(0, emb_size, 2)* math.log(10000) / emb_size)
        pos = torch.arange(0, maxlen).reshape(maxlen, 1)
        pos_embedding = torch.zeros((maxlen, emb_size))
        pos_embedding[:, 0::2] = torch.sin(pos * den)
        pos_embedding[:, 1::2] = torch.cos(pos * den)
        pos_embedding = pos_embedding.unsqueeze(-2)

        self.dropout = nn.Dropout(dropout)
        self.register_buffer("pos_embedding", pos_embedding)

    def forward(self, token_embedding: Tensor):
        return self.dropout(token_embedding + self.pos_embedding[:token_embedding.size(0), :])


class TokenEmbedding(nn.Module):
    def __init__(self, vocab_size: int, emb_size):
        super(TokenEmbedding, self).__init__()
        self.embedding = nn.Embedding(vocab_size, emb_size)
        self.emb_size = emb_size

    def forward(self, tokens: Tensor):
        return self.embedding(tokens.long()) * math.sqrt(self.emb_size)

class MultiNet(nn.Module):
    def __init__(self,
                 num_encoder_layers: int,
                 num_decoder_layers: int,
                 emb_size: int,
                 nhead: int,
                 src_vocab_size: int,
                 tgt_vocab_size: int,
                 dim_feedforward: int = 512,
                 dropout: float = 0.1):
        super(MultiNet, self).__init__()
        self.transformer = Transformer(d_model=emb_size,
                                       nhead=nhead,
                                       num_encoder_layers=num_encoder_layers,
                                       num_decoder_layers=num_decoder_layers,
                                       dim_feedforward=dim_feedforward,
                                       dropout=dropout)
        self.generator = nn.Linear(emb_size, tgt_vocab_size)
        self.src_tok_emb = TokenEmbedding(src_vocab_size, emb_size)
        self.tgt_tok_emb = TokenEmbedding(tgt_vocab_size, emb_size)
        self.positional_encoding = PositionalEncoding(
            emb_size, dropout=dropout)

    def forward(self,
                src: Tensor,
                trg: Tensor,
                src_mask: Tensor,
                tgt_mask: Tensor,
                src_padding_mask: Tensor,
                tgt_padding_mask: Tensor,
                memory_key_padding_mask: Tensor):
        src_emb = self.positional_encoding(self.src_tok_emb(src))
        tgt_emb = self.positional_encoding(self.tgt_tok_emb(trg))
        outs = self.transformer(src_emb, tgt_emb, src_mask, tgt_mask, None,
                                src_padding_mask, tgt_padding_mask, memory_key_padding_mask)
        return self.generator(outs)

    def encode(self, src: Tensor, src_mask: Tensor):
        return self.transformer.encoder(self.positional_encoding(
                            self.src_tok_emb(src)), src_mask)

    def decode(self, tgt: Tensor, memory: Tensor, tgt_mask: Tensor):
        return self.transformer.decoder(self.positional_encoding(
                          self.tgt_tok_emb(tgt)), memory,
                          tgt_mask)



SRC_VOCAB_SIZE = len(vocab_transform[SRC_LANGUAGE])
TGT_VOCAB_SIZE = len(vocab_transform[TGT_LANGUAGE])
EMB_SIZE = 512
NHEAD = 8
FFN_HID_DIM = 512
NUM_ENCODER_LAYERS = 3
NUM_DECODER_LAYERS = 3
```

Перепишем функции для тренировки и оценки под специфику текущей модели

```python
def train_seq2seq(
    data,
    model: nn.Module,
    optimizer: torch.optim.Optimizer,
    loss_fn,
    device,
):
    src, tgt = data
    src, tgt = (
        src.to(device),
        tgt.to(device),
    )


    tgt_input = tgt[:-1, :]
    src_mask, tgt_mask, src_padding_mask, tgt_padding_mask = create_mask(src, tgt_input)

    # zero the parameter gradients
    optimizer.zero_grad()

    # forward pass and loss calculation
    logits = model(src, tgt_input, src_mask, tgt_mask,src_padding_mask, tgt_padding_mask, src_padding_mask)

    tgt_out = tgt[1:, :].long()

    return loss_fn(logits.reshape(-1, logits.shape[-1]), tgt_out.reshape(-1))


def eval_seq2seq(
    test_loader: DataLoader,
    model: nn.Module,
    loss_fn,
    device,
):

    val_loss = 0.0
    with torch.no_grad():
        model.eval()  # evaluation mode
        for src, tgt in test_loader:
            src, tgt = (
                src.to(device),
                tgt.to(device),
            )

            tgt_input = tgt[:-1, :]
            src_mask, tgt_mask, src_padding_mask, tgt_padding_mask = create_mask(src, tgt_input)


            logits = model(src, tgt_input, src_mask, tgt_mask,src_padding_mask, tgt_padding_mask, src_padding_mask)
            tgt_out = tgt[1:, :].long()


            val_loss += loss_fn(logits.reshape(-1, logits.shape[-1]), tgt_out.reshape(-1)).item()

    return val_loss, 0



def get_multi_net():
  model = MultiNet(NUM_ENCODER_LAYERS, NUM_DECODER_LAYERS, EMB_SIZE,
                                 NHEAD, SRC_VOCAB_SIZE, TGT_VOCAB_SIZE, FFN_HID_DIM)

  for p in model.parameters():
    if p.dim() > 1:
        nn.init.xavier_uniform_(p)

  return model
```

Модели: генератор и дискриминатор

```python
LATENT_VECTOR_SIZE = 100
FIXED_NOISE = torch.randn(64, LATENT_VECTOR_SIZE, 1, 1, device=DEVICE)

def weights_init(m):
    classname = m.__class__.__name__
    if classname.find("Conv") != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
    elif classname.find("BatchNorm") != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)

class LfwGenerator(nn.Module):
    def __init__(self, ngpu, nc=3, nz=LATENT_VECTOR_SIZE, ngf=64):
        super(LfwGenerator, self).__init__()
        self.ngpu = ngpu
        self.main = nn.Sequential(
            nn.ConvTranspose2d( nz, ngf * 8, 4, 1, 0, bias=False),
            nn.BatchNorm2d(ngf * 8),
            nn.ReLU(True),
            nn.ConvTranspose2d(ngf * 8, ngf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 4),
            nn.ReLU(True),
            nn.ConvTranspose2d( ngf * 4, ngf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 2),
            nn.ReLU(True),
            nn.ConvTranspose2d( ngf * 2, ngf, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf),
            nn.ReLU(True),
            nn.ConvTranspose2d( ngf, nc, 4, 2, 1, bias=False),
            nn.Tanh()
        )

    def forward(self, input):
        return self.main(input)


class LfwDiscriminator(nn.Module):
    def __init__(self, ngpu, nc=3, ndf=64):
        super(LfwDiscriminator, self).__init__()
        self.ngpu = ngpu
        self.main = nn.Sequential(
            nn.Conv2d(nc, ndf, 4, 2, 1, bias=False),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(ndf, ndf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ndf * 2),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(ndf * 2, ndf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ndf * 4),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(ndf * 4, ndf * 8, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ndf * 8),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(ndf * 8, 1, 4, 1, 0, bias=False),
            nn.Sigmoid()
        )

    def forward(self, input):
        return self.main(input)


def get_lfw_generator():
    ngpu = 1 if DEVICE.type == "cuda" else 0

    lfw_gen = LfwGenerator(ngpu).to(DEVICE)

    if (DEVICE.type == "cuda") and (ngpu > 1):
        lfw_gen = nn.DataParallel(lfw_gen, list(range(ngpu)))

    lfw_gen.apply(weights_init)

    return lfw_gen

def get_lfw_discriminator():
    ngpu = 1 if DEVICE.type == "cuda" else 0

    lfw_disc = LfwDiscriminator(ngpu).to(DEVICE)

    if (DEVICE.type == "cuda") and (ngpu > 1):
        lfw_disc = nn.DataParallel(lfw_disc, list(range(ngpu)))

    lfw_disc.apply(weights_init)

    return lfw_disc
```

Перепишем функции тренировки и оценки под архитектуру DCGAN

```python
def train_gan(
    data,
    generator: nn.Module,
    discriminator: nn.Module,
    optimizer_g: torch.optim.Optimizer,
    optimizer_d: torch.optim.Optimizer,
    loss_fn,
    device,
):
    real_label = 1
    fake_label = 0

    optimizer_g.zero_grad()
    optimizer_d.zero_grad()

    # Train discriminator on real data
    discriminator.zero_grad()
    discriminator.train()
    real_imgs = data[0].to(device)
    b_size = real_imgs.size(0)
    label = torch.full((b_size,), real_label, dtype=torch.float, device=device)
    output = discriminator(real_imgs).view(-1)

    err_disc_real = loss_fn(output, label)

    # Train discriminator on fake data
    noise = torch.randn(b_size, LATENT_VECTOR_SIZE, 1, 1, device=device)
    fake_imgs = generator(noise)
    label = torch.full((b_size,), fake_label, dtype=torch.float, device=device)
    label.fill_(fake_label)
    output = discriminator(fake_imgs.detach()).view(-1)

    err_disc_fake = loss_fn(output[:label.shape[0]], label)

    err_disc = err_disc_real + err_disc_fake

    # Train generator
    generator.zero_grad()
    generator.train()
    label = torch.full((b_size,), real_label, dtype=torch.float, device=device)
    label.fill_(real_label)
    output = discriminator(fake_imgs).view(-1)

    err_gen = loss_fn(output[:label.shape[0]], label)

    return err_gen, err_disc


def eval_gan(
    test_loader: DataLoader,
    generator: nn.Module,
    discriminator: nn.Module,
    fixed_noise,
    loss_fn,
    device,
):
    real_label = 1
    fake_label = 0

    val_loss_g = 0.0
    val_loss_d = 0.0
    with torch.no_grad():
        generator.eval()  # evaluation mode
        discriminator.eval()  # evaluation mode
        for data in test_loader:

            # Evaluate discriminator on real data
            real_imgs = data[0].to(device)
            b_size = real_imgs.size(0)
            label = torch.full((b_size,), real_label, dtype=torch.float, device=device)
            output = discriminator(real_imgs).view(-1)

            err_disc_real = loss_fn(output, label)


            # Evaluate discriminator on fake data
            fake_imgs = generator(fixed_noise)
            label.fill_(fake_label)
            output = discriminator(fake_imgs.detach()).view(-1)

            err_disc_fake = loss_fn(output[:label.shape[0]], label)

            err_disc = err_disc_real + err_disc_fake


            # Evaluate generator
            label.fill_(real_label)
            output = discriminator(fake_imgs).view(-1)

            err_gen = loss_fn(output[:label.shape[0]], label)


            val_loss_g += err_gen.item()
            val_loss_d += err_disc.item()

    return val_loss_g, val_loss_d


def train_eval_loop_gan(
    generator: nn.Module,
    discriminator: nn.Module,
    optimizer_g: torch.optim.Optimizer,
    optimizer_d: torch.optim.Optimizer,
    fixed_noise,
    loss_fn,
    train_loader: DataLoader,
    test_loader: DataLoader,
    train_fn,
    eval_fn,
    configs: dict,
    device,
    log_frequency: int,
    verbose: bool,
    include_starting_point: bool,
    leave_tqdm: bool,
    epochs=1,
):
    loss_g_logs = []
    val_loss_g_logs = []
    loss_d_logs = []
    val_loss_d_logs = []

    time_logs = []
    iterations_logs = []

    torch.cuda.synchronize()
    start_time = time.time()

    iterations = 0

    if include_starting_point:
        # logs
        iterations_logs.append(iterations)
        torch.cuda.synchronize()
        time_logs.append(0)

        # evaluation
        val_loss_g, val_loss_d = eval_fn(test_loader, generator, discriminator, fixed_noise, loss_fn, device=device)

        val_loss_g_logs.append(val_loss_g)
        val_loss_d_logs.append(val_loss_d)


    # iterating over epochs
    for epoch in range(epochs):
        # training loop description
        train_loop = tqdm(
            enumerate(train_loader),
            total=len(train_loader),
            desc=f"Epoch {epoch+1}",
            disable=not verbose,
            leave=leave_tqdm,
        )
        train_loss_g = 0.0
        train_loss_d = 0.0
        # iterate over dataset
        for i, data in train_loop:

            loss_g, loss_d = train_fn(data, generator, discriminator, optimizer_g, optimizer_d, loss_fn, device=device)

            # backward pass
            loss_g.backward()
            loss_d.backward()

            # gradient clipping
            if configs.get("clipping") and configs["clipping"]["apply"]:
                for model in [generator, discriminator]:
                    nn.utils.clip_grad_norm_(
                        model.parameters(),
                        max_norm=configs["clipping"]["threshold"],
                        norm_type=configs["clipping"]["norm_type"],
                    )

            # optimizer run
            optimizer_g.step()
            optimizer_d.step()

            train_loss_g += loss_g.item()
            train_loss_d += loss_d.item()

            iterations += 1
            if iterations % log_frequency != 0:
                continue

            # logs
            iterations_logs.append(iterations)
            loss_g_logs.append(loss_g.item())
            loss_d_logs.append(loss_d.item())
            torch.cuda.synchronize()
            time_logs.append(time.time() - start_time)

            # evaluation
            val_loss_g, val_loss_d = eval_fn(test_loader, generator, discriminator, fixed_noise, loss_fn, device=device)

            val_loss_g_logs.append(val_loss_g)
            val_loss_d_logs.append(val_loss_d)

            train_loop.set_postfix({"val loss g": val_loss_g, "val loss d": val_loss_d})

    return val_loss_g_logs, val_loss_d_logs, time_logs, iterations_logs
```

```python
def methods_run_gan(
    get_g,
    get_d,
    loss_fn,
    train_loader_lambda,
    test_loader: DataLoader,
    to_compare: dict[str, dict],
    fixed_noise = FIXED_NOISE,
    epochs: int = 5,
    log_frequency: float = 50,
    seed: int = 420,
    verbose: bool = False,
    train_fn=train_gan,
    eval_fn=eval_gan,
    include_starting_point: bool = False,
    device=DEVICE,
    leave_tqdm: bool = False,
) -> list[dict]:

    comparison_dicts = []

    for name, configs in to_compare.items():
        if verbose:
            print(f"Running {name.upper()}...")

        torch.manual_seed(seed)
        torch.cuda.manual_seed(seed)
        np.random.seed(seed)

        generator = get_g().to(device)
        discriminator = get_d().to(device)
        optimizer_g = configs["optimizer_lambda"](generator)
        optimizer_d = configs["optimizer_lambda"](discriminator)
        train_loader = train_loader_lambda(configs)

        if log_frequency <= 0:
            literal_log_frequency = len(train_loader)
        elif log_frequency < 1:
            literal_log_frequency = int(len(train_loader) * log_frequency)
        else:
            literal_log_frequency = int(log_frequency)

        val_loss_g_logs, val_loss_d_logs, time_logs, iterations_logs = train_eval_loop_gan(
            generator,
            discriminator,
            optimizer_g,
            optimizer_d,
            fixed_noise,
            loss_fn,
            train_fn=train_fn,
            eval_fn=eval_fn,
            epochs=epochs,
            train_loader=train_loader,
            test_loader=test_loader,
            configs=configs,
            device=device,
            log_frequency=literal_log_frequency,
            verbose=verbose,
            include_starting_point=include_starting_point,
            leave_tqdm=leave_tqdm,
        )

        comparison_dicts.append(
            {
                "name": name,
                "Test generator loss": val_loss_g_logs,
                "Test discriminator loss": val_loss_d_logs,
                "Time": time_logs,
                "Iterations": iterations_logs,
            }
        )

    if verbose:
        print("Done!")
    return comparison_dicts


def methods_plot_comparison_gan(title: str, comparison_dicts: list[dict]):

    plots = [
            (
                {"name": "Iterations", "label": "Iterations"},
                {"name": "Test generator loss", "label": "Loss"},
            ),
            (
                {"name": "Time", "label": "Running time, seconds"},
                {"name": "Test generator loss", "label": "Loss"},
            ),

            (
                {"name": "Iterations", "label": "Iterations"},
                {"name": "Test discriminator loss", "label": "Loss"},
            ),
            (
                {"name": "Time", "label": "Running time, seconds"},
                {"name": "Test discriminator loss", "label": "Loss"},
            ),
        ]

    draw_method_plots(
        comparison_dicts,
        title,
        plots,
        plot_width=7,
        plot_height=4,
        row_plots=2,
    )
```
